```python
#Linked list implementation of Stack in Python

import math

class Node:
  def __init__(self,data):
    self.data=data
    self.next=None

class MyStack:
  def __init__(self):
    self.head=None
    self.sz=0

  def push(self,x):
    temp=Node(x)
    temp.next=self.head
    self.head=temp
    self.sz=self.sz+1

  def size(self):
    return self.sz

  def peek(self):
    if self.head==None:
      return math.inf
    return self.head.data

  def pop(self):
    if self.head==None:
      return math.inf
    res=self.head.data
    self.head=self.head.next
    self.sz=self.sz-1
    return res
```

```python
s=MyStack()
s.push(10)
s.push(20)
s.push(30)
print(s.pop())
print(s.peek())
print(s.size())
```

```python
# Check for balanced parenthesis



def is_matching(a,b):
  if (a=='(' and b==')') or (a=='{' and b=='}') or (a=='[' and b==']'):
    return True
  return False

def is_balanced(expr):
  stack=[]
  for x in (expr):
    if x in ('(','{','['):
      stack.append(x)
    else:
      if not stack:
        return False
      elif is_matching(stack[-1],x)==False:
        return False
      else:
        stack.pop()

  if stack:
    return False
  else:
    return True


is_balanced('[{([])}]')
```
