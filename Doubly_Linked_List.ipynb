```python
# DOUBLY LINKED LIST

class Node:
  def __init__(self,data):
    self.data=data
    self.prev=None
    self.next=None
```

```python
def printList(head):
  curr=head
  while curr != None:
    print(curr.data,end=" ")
    curr=curr.next
```

```python
head=Node(10)
temp1=Node(20)
temp2=Node(30)

head.next=temp1
temp1.prev=head
temp1.next=temp2
temp2.prev=temp1

printList(head)
```

```python
def insertBegin(head,data):
  temp=Node(data)
  if head !=None:
    head.prev=temp
  temp.next=head
  return temp

head=insertBegin(head,10)
head=insertBegin(head,20)

printList(head)
```

```python
def insertEnd(head,data):
  temp=Node(data)
  if head is None:
    return temp

  curr=head
  while curr.next != None:
    curr=curr.next
  curr.next=temp
  temp.prev=curr
  return head

head=insertEnd(head,30)
head=insertEnd(head,40)

printList(head)
```

```python
def delHead(head):
  if head ==None:
    return None
  if head.next==None:
    return None
  else:
    head=head.next
    head.prev=None
    return head

head=delHead(head)
printList(head)
```

```python
def delEnd(head):

  if head==None or head.next==None:
    return None

  curr=head
  while curr.next.next!=None:
    curr=curr.next

  curr.next=None
  return head

head=delEnd(head)

printList(head)
```

```python
def reverseDLL(head):
  if head is None or head.next is None:
    return head

  curr = head
  prev = None # 'prev' will track the previous node in the reversed list, and ultimately become the new head

  while curr is not None:
    # Store the original next node before its pointer is modified
    next_node_to_process = curr.next

    # Swap curr.prev and curr.next pointers
    curr.prev = curr.next # The new 'prev' for curr is its original 'next'
    curr.next = prev      # The new 'next' for curr is the 'prev' node from the *previous* iteration

    # 'prev' now becomes the current node, as it's the most recently reversed node
    # and will be the 'next' for the node that comes *before* it in the original list
    prev = curr

    # Move to the next node in the ORIGINAL list
    curr = next_node_to_process

  return prev # 'prev' holds the new head of the reversed list (which was the original tail)

head=reverseDLL(head)
printList(head)
```
