```python
class Node:
  def __init__(self,k):
    self.key = k
    self.next = None

temp1=Node(10)
temp2=Node(20)
temp3=Node(30)

temp1.next=temp2
temp2.next=temp3
head=temp1



```

```python
def printList(head):
  temp=head
  while(temp!=None):
    print(temp.key)
    temp=temp.next

printList(head)
```

```python
def search(head,key):
  pos=1
  curr=head
  while(curr!=None):
    if curr.key==key:
      return pos
    pos+=1
    curr=curr.next
  return -1

print(search(head,30))
```

```python
# Insert at the beginning of Linked List
def insertbegin(head,key):
  temp=Node(key)
  temp.next=head
  return temp

head=insertbegin(head,5)
printList(head)
```

```python
# Insert at the end of the Linked List

def insertend(head,key):
  if head==None:
    return Node(key)
  temp=Node(key)
  curr=head
  while(curr.next!=None):
    curr=curr.next
  curr.next=temp
  return head

head=insertend(head,40)
printList(head)
```

```python
#Insert at a Position
def insertPos(head,pos,data):
  temp=Node(data)

  if pos==1:
    temp.next=head
    return temp
  curr=head
  for i in range(pos-2):
    curr=curr.next
    if curr==None:
      return head
  temp.next=curr.next
  curr.next=temp
  return head

head=insertPos(head,3,25)
printList(head)
```

```python
#Delete First Node of Linked List

def delete_first(head):
  if head==None:
    return None
  else:
    return head.next

head=delete_first(head)
printList(head)
```

```python
# Delete Last Node of Linked List

def delete_last(head):
  if head==None:
    return None
  if head.next==None:
    return None
  curr=head
  while(curr.next.next!=None):
    curr=curr.next
  curr.next=None
  return head

head=delete_last(head)
printList(head)
```

```python
# Delete the node with only pointer given to it
def deleteNode(ptr):
  temp=ptr.next
  ptr.data=temp.data
  ptr.next=temp.next
```

```python
# Sorted Insert Linked List in Python

def sorted_insert(head,value):
  temp=Node(value)
  curr=head
  while(curr.next!=None and curr.next.key<value):
    curr=curr.next
  temp.next=curr.next
  curr.next=temp
  return head

head=sorted_insert(head,25)
printList(head)
```

```python
# Middle of a Linked List

def printMiddle(head):
  if head == None:
    return None
  count=0
  curr=head
  while curr:
    curr=curr.next
    count+=1
  curr=head
  for i in range(count//2):
    curr=curr.next
  print(curr.key)

printMiddle(head)
```

```python
# Middle of a Linked List efficient

def printMiddle_efficient(head):
  if head ==None:
    return None

  slow=head
  fast=head
  while fast!=None and fast.next!=None:
    slow=slow.next
    fast=fast.next.next
  print(slow.key)

printMiddle_efficient(head)
```

```python
# Nth Node from the end of the Linked List

def printNthFromEnd(head, n):
  if head == None:
    return None
  len=0
  curr=head
  while curr:
    curr=curr.next
    len+=1
  if len<n:
    return None
  curr=head
  for i in range(1,len-n+1):
    curr=curr.next
  print(curr.key)

printNthFromEnd(head,3)
```

```python
# Nth Node from the end of the Linked List using 2 pointers

def printNthFromEnd2(head,n):
  if head == None:
    return None
  first=head
  second=head
  for i in range(n):
    if first==None:
      return None
    first=first.next
  while first:
    first=first.next
    second=second.next
  print(second.key)

printNthFromEnd2(head,2)
```

```python
# Remove Duplicates from a Sorted List
def removeDup(head):
  curr=head
  while curr!=None and curr.next!=None:
    if curr.data==curr.next.data:
      curr.next=curr.next.next
    else:
      curr=curr.next
```

```python
# Reverse a Linked List in Python
def reverse_linked_list(head):
  curr=head
  prev=None
  while curr is not None:
    next=curr.next
    curr.next=prev
    prev=curr
    curr=next
  return prev

new_head = reverse_linked_list(head)
printList(new_head)



```

```python
# Reverse Recursive a Linked List

def reverseList(head):
  if head==None or head.next==None:
    return head
  rest_head=reverseList(head.next)
  head.next.next=head
  head.next=None
  return rest_head

new_head=reverseList(head)
printList(new_head)
```

```python
# DOUBLY LINKED LIST

class Node:
  def __init__(self,data):
    self.data=data
    self.prev=None
    self.next=None

```

```python
head=Node(10)
temp1=Node(20)
temp2=Node(30)

head.next=temp1
temp1.prev=head
temp1.next=temp2
temp2.prev=temp1
```

```python
def insertBegin(head,data):
  temp=Node(data)
  if head !=None:
    head.prev=temp
  temp.next=head
  return temp

head=insertBegin(head,10)
head=insertBegin(head,20)

printList(head)
```
