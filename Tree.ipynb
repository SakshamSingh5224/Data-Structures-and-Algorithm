```python
class Node:
  def __init__(self,k):
    self.left=None
    self.right=None
    self.key=k
```

```python
root=Node(10)
root.left=Node(20)
root.right=Node(30)
root.right.left=Node(40)
root.right.right=Node(50)
```

```python
def inorder(root):
  if root!=None:
    inorder(root.left)
    print(root.key)
    inorder(root.right)

inorder(root)
```

```python
def preorder(root):
  if root!=None:
    print(root.key)
    preorder(root.left)
    preorder(root.right)

preorder(root)
```

```python
def postorder(root):
  if root!=None:
    postorder(root.left)
    postorder(root.right)
    print(root.key)

postorder(root)
```

```python
def height(root):
  if root==None:
    return 0
  else:
    lh=height(root.left)
    rh=height(root.right)

    return max(lh,rh)+1

height(root)
# Shorter implementation
# def height(root):
#   if root==None:
#     return 0
#   else:
#     return max(height(root.left),height(root.right))+1
```

```python
# Print nodes at distance K

def printKDist(root,k):
  if root is None:
    return
  if k==0:
    print(root.key,end=" ")

  else:
    printKDist(root.left,k-1)
    printKDist(root.right,k-1)

printKDist(root,2)

```

```python
# Level Order Traversal

#Naive method

# h=height(root)
# for k in range(h):
#   printDist(root,k)

# Time Complexity : O(n+nh)=O(nh)

from collections import deque

def printLevelOrder(root):
  if root is None:
    return
  q=deque()
  q.append(root)
  while len(q)>0:
    node=q.popleft()
    print(node.key,end=" ")
    if node.left is not None:
      q.append(node.left)
    if node.right is not None:
      q.append(node.right)

printLevelOrder(root)

# Time Complexity : Theta(N)
# Auxililary Space: O(N)



```

```python
# Size of the Binary Tree

def treeSize(root):
  if root is None:
    return 0
  else:
    ls=treeSize(root.left)
    rs=treeSize(root.right)
    return ls + rs +1
    # Shorter Implementation return treeSize(root.left)+treeSize(root.right)+1
treeSize(root)

# Time Complexity : Theta(N)
# Auxililary Space: Theta(h)
```

```python
# Maximum in a Binary Tree in Python
import math
def getMax(root):
  if root is None:
    return -math.inf

  else:
    lm=getMax(root.left)
    rm=getMax(root.right)

    return max(root.key,lm ,rm)

getMax(root)

# Time Complexity : Theta(N)
# Auxililary Space: Theta(h)
```
